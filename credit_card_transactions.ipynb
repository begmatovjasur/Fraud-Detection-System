```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay 
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
X = df.drop(columns=(['time','class']))
y = df['class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
model = LogisticRegression(class_weight='balanced')
model.fit(X_train,y_train)
```

```python
prediction = model.predict(X_test)
acc = accuracy_score(prediction,y_test)
print(f"Accuracy {acc}")
```

```python
ConfusionMatrixDisplay.from_estimator(model,X_test,y_test, cmap='Blues')
plt.show()
```

```python
model_rf = RandomForestClassifier(n_estimators=100, class_weight='balanced')
model_rf.fit(X_train, y_train)
```

```python
prediction_rf = model_rf.predict(X_test)
acc_rf = accuracy_score(prediction_rf,y_test)
print(f"Accuracy: {acc_rf}")
```

```python
ConfusionMatrixDisplay.from_estimator(model_rf,X_test,y_test, cmap='Blues')
plt.show()
```

```python
y_prob = model_rf.predict_proba(X_test)[:,1]
threshold = 0.3
y_hat = (y_prob > threshold).astype(int)
ConfusionMatrixDisplay.from_predictions(y_hat,y_test, cmap='Blues')
plt.show()
```

### SHAP Analysis

Random Forest "Qora quti" (Black-box) model bo'lganligi sababli, uning qarorlarini biznes tili bilan tushuntirish uchun **SHAP (SHapley Additive exPlanations)** kutubxonasidan foydalandik. Yuqoridagi Summary Plot modelimizning arxitekturasini to'liq ochib beradi.

**Asosiy xulosalar:**
1. **Eng muhim omillar:** `V12`, `V14`, `V4` va `V11` ustunlari firibgarlikni aniqlashda eng katta vaznga ega. Model tranzaksiyani bloklashdan oldin birinchi navbatda shu ko'rsatkichlarni tekshiradi.
2. **Teskari korrelyatsiya (V12 va V14):** Grafikdan ko'rinib turibdiki, ushbu o'zgaruvchilarning qiymati keskin pasayishi (ko'k nuqtalar) tranzaksiyaning firibgarlik ehtimolini (SHAP value > 0) yuqori darajada oshiradi.
3. **To'g'ri korrelyatsiya (V4 va V11):** Aksincha, bu ko'rsatkichlarning oshib ketishi (qizil nuqtalar) modelga shubhali faollik haqida kuchli signal yuboradi.

Bu tahlil bank xavfsizlik bo'limiga nafaqat firibgarlarni ushlashda, balki ularning qanday mantiq asosida harakat qilayotganini tushunishda yordam beradi.

```python
from xgboost import XGBClassifier
from sklearn.metrics import classification_report
```

```python
weight = y_train.value_counts()[0] / y_train.value_counts()[1]
model_xgb = XGBClassifier(
    n_estimators=100, 
    scale_pos_weight=weight, 
    random_state=42,
    use_label_encoder=False,
    eval_metric='logloss'
)
model_xgb.fit(X_train, y_train)
```
